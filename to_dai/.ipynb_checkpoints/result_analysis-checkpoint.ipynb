```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pred_csv_path = 'no2_union_with_o3_MaryleboneRoad.csv'
df_res = pd.read_csv(pred_csv_path)
```

```python
df_res['error'] = df_res['no2'] -  df_res['no2_predict'] 
```

```python
fig, ax = plt.subplots()
sns.distplot(df_res['error'], ax=ax)
ax.set_xlim(-60,60)
plt.title("no2 error distribution-MY1 in no2 union model")
#ax.set_xticks(range(1,32))
```

```python
fig, ax = plt.subplots()
sns.distplot(df_res['no2'], ax=ax)
#ax.set_xlim(-60,60)
plt.title("no2 ground truth distribution-MY1")
#ax.set_xticks(range(1,32))
```
