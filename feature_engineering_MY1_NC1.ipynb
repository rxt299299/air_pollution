```python
air_pollution = 'no2'

# from date unix extract year
def extract_year(row):
    try:
        date_unix = int(row['date_unix'])
        return datetime.utcfromtimestamp(date_unix).year
    except:
        return 0

def fill_continuous_missValues(air_pollution,
                               continuous_variables,
                               df):    
    #change the continuous features
    for feature in ['ap_lat', 'ap_long', 'met_lat', 'met_long']:
        if feature in continuous_variables:
            continuous_variables.remove(feature)

    for feature in ['o3']:
        if feature not in continuous_variables:
            continuous_variables.append(feature)
        
    print (continuous_variables)
    #add column year
    df['year'] = df.apply(lambda row : extract_year(row), axis=1)
    df = df[df['year']>0]

    for colname in continuous_variables:
        df[colname] = df[colname].fillna(df[colname].median())
        
    return df
```

```python
df_MY1 = fill_continuous_missValues(air_pollution,
                                    continuous_variables+['tc'],
                                    df_MY1[selected_columns_MY1])
df_KC1 = fill_continuous_missValues(air_pollution,
                                    continuous_variables,
                                    df_KC1[selected_columns_KC1])
```

```python
 df_MY1.to_csv(('D:/work/污染项目/data-overall' \
                     '-update-0120/processed_MaryleboneRoad.csv'), index=False)
    
    
df_KC1.to_csv(('D:/work/污染项目/data-overall' \
                 '-update-0120/processed_NorthKensington.csv'), index=False)
```
